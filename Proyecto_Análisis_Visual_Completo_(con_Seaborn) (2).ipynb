# Proyecto: Análisis Visual Completo (con Seaborn)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/medallistas_panamericanos_CA.csv")
```

```python
df.info()
```

```python
# Estilo general
sns.set_theme(style="whitegrid")
```

```python
# 1) Conteo de medallas por país

plt.figure(figsize=(10,6))
sns.countplot(data=df, x="País", palette="viridis")
plt.title("Cantidad de Medallas por País", color="pink")
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig("medallas_pais.png", dpi=300)
plt.show()
```

```python
# 2) Conteo por tipo de medalla (Gold, Silver, Bronze)

plt.figure(figsize=(8,6))
sns.countplot(data=df, x="Medalla", palette="pastel", order=df["Medalla"].value_counts().index)
plt.title("Distribución de Tipos de Medalla")
plt.tight_layout()
plt.savefig("tipos_medalla.png", dpi=300)
plt.show()
```

```python
# 3) Boxplot del Año por Medalla

plt.figure(figsize=(8,6))
sns.boxplot(data=df, x="Medalla", y="Año", palette="Set2")
plt.title("Distribución del Año por Tipo de Medalla (Boxplot)")
plt.tight_layout()
plt.savefig("boxplot_anio_medalla.png", dpi=300)
plt.show()
```

```python
# 4) Nuevo Heatmap – Frecuencia País vs Medalla

tabla = df.pivot_table(index="País", columns="Medalla", aggfunc="size", fill_value=0)

plt.figure(figsize=(10,6))
sns.heatmap(tabla, annot=True, cmap="YlOrBr")
plt.title("Frecuencia de Medallas por País (Heatmap)",
          color="red", fontsize=14, weight="bold")
plt.xlabel("Medalla")
plt.ylabel("País")
plt.tight_layout()
plt.savefig("heatmap_pais_medalla.png", dpi=300)
plt.show()
```

```python
# 5) Violinplot Año vs Deporte – paleta "rocket"
plt.figure(figsize=(12,6))
sns.violinplot(data=df, x="Deporte", y="Año", palette="rocket")
plt.title("Distribución del Año por Deporte (Violinplot)")
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig("violinplot_anio_deporte.png", dpi=300)
plt.show()
```
